```python
from duckduckgo_search import DDGS
import requests, os, shutil

def download_images(keyword, save_dir, max_images=100):
    os.makedirs(save_dir, exist_ok=True)
    count = 0

    with DDGS() as ddgs:
        results = ddgs.images(keyword, max_results=max_images)
        
        for i, r in enumerate(results):
            try:
                url = r["image"]
                img = requests.get(url, timeout=5)
                
                if img.status_code == 200:
                    with open(f"{save_dir}/{keyword.replace(' ','_')}_{i}.jpg", "wb") as f:
                        f.write(img.content)
                    count += 1
            except:
                pass
    
    print(f"âœ” Downloaded {count} images for: {keyword}")


# -------- Download Healthy Soybean Leaves ----------
healthy_keywords = [
    "healthy soybean leaf",
    "green soybean leaf close up",
    "soybean healthy leaves"
]

for kw in healthy_keywords:
    download_images(kw, "soybean/healthy", max_images=80)

# -------- Download Diseased Soybean Leaves ----------
diseased_keywords = [
    "soybean leaf disease",
    "soybean leaf spot",
    "soybean rust",
    "soybean bacterial blight",
    "soybean mosaic virus",
    "soybean downy mildew",
    "soybean frogeye leaf spot"
]

for kw in diseased_keywords:
    download_images(kw, "soybean/diseased", max_images=80)


print("\nðŸŽ¯ DONE! All healthy & diseased images downloaded successfully.")
```
